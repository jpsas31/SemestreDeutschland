```python
##authentication
import json
import pandas as pd
client_secrets_file = "secret.json"
file= open(client_secrets_file)
jsonSecret= json.load(file)
key=jsonSecret['key']
user='nba'
def last_fm(payload):
    headers = {'user-agent':user}
    url='http://ws.audioscrobbler.com/2.0/?'
    
  
    payload['api_key']=key
    payload['format'] = 'json'
    res= requests.get(url, headers=headers, params=payload)

    return res.json()
    

res=last_fm({
    'method':'artist.getinfo',
    'artist':'Mac demarco'
})
artists=[]
for artist in res['artist']['similar'].values():
    artists.append(artist)
# print(artists)
dt= pd.DataFrame(artists[0])
dt



# dt
```
