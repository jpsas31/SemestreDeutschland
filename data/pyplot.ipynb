```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
%matplotlib inline
plt.style.use('dark_background')
```

```python
#Task one
y= np.array([x**2 for x in range(1,7)])
x = np.array([x for x in range(1,7  )])
fig, axs = plt.subplots()
axs.axhline(20,color='r')
axs.scatter(x,y)
axs.set_ylabel('$x^2$')
plt.title('A simple diagramm')
axs.annotate('This one',xy=(3,9), xytext=(4,12),arrowprops=dict({'facecolor':'white'}))
plt.show()
```
